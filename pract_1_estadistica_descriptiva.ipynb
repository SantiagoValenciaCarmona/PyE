**Cargas bases a R**

Hay al menos tres formas de cargar los datos:

- con extensión xlsx (Excel normal)
- con extensión csv (Excel guardados con csv separados por coma)
- con extensión txt (datos que provienen de bloq de notas)

* **Para xlsx**

Primero se debe cargar la librería readxl como se muestra a continuación.

***Nota:***
Para bases con extensión csv y txt no se necesita cargar la librería readxl

```r
install.packages("readxl")
library(readxl)
```

```r
# Cargar la base
# Nota: antes de ejecutrar las siguientes líneas, debes verificar que la base este en la carpeta de la izquierda (en los archivos)

datos=read_excel("/content/Datos_1.xlsx")   # sheet = NULL
head(datos)
attach(datos)
```

* **Para csv**

```r
# Nota: antes de ejecutar la línea, verificar que en la base los datos estén separados por coma.
datos1<-read.csv("/content/forestfires.csv",sep=",",dec = ".",
                 header = T)  # Se coloca header=T, cuando la base tiene nombres en las columnas, cuando no los tiene se coloca header=F
head(datos1)
attach(datos1)
```

* **Para txt**

```r
datos2=read.table("/content/iris.txt", header = F, sep = ",") # se coloca header = FALSE, porque la base no tiene nombres en las columnas y se coloca sep = ",", porque los datos están separados por coma en la base.
head(datos2)
attach(datos2)
```
